```python
# Function to all features of product
def get_title(soup):

    try:
        title = Soup1.find("span", attrs={"id":'productTitle'}).text.strip()
    except AttributeError:
        title = ""
    return title

def get_price(soup):
    try:
        price = Soup1.find("span", attrs={"class":'a-offscreen'}).string.strip()
    except AttributeError:
        try:
            price = Soup1.find("span", attrs={"class":'a-size-small aok-offscreen'}).string.strip()
        except:
            price = "Out of Stock"
    return price

def get_rating(soup):
    try:
        rating = Soup1.find("i", attrs={'class':'a-icon a-icon-star a-star-4-5'}).string.strip()   
    except AttributeError:
        try:
            rating = Soup1.find("span", attrs={'class':'a-icon-alt'}).string.strip()
        except:
            rating = ""
    return rating

def get_review_count(soup):
    try:
        review_count = Soup1.find("span", attrs={'id':'acrCustomerReviewText'}).string.strip()
    except AttributeError:
        review_count = ""
    return review_count

def get_availability(soup):
    try:
        available = Soup1.find("div", attrs={'id':'availability'})
        available = available.find("span").string.strip()

    except AttributeError:
        available = "Not Available"	

    return available

def get_brand(soup):
    try:
        brand = Soup1.find("tr", attrs={"class":'a-spacing-small po-brand'}).find("span", attrs={"class": "a-size-base po-break-word"}).string.strip()
    except AttributeError:
        brand = ""           
    return brand

def get_number_of_cells(soup):
    try:
        noOfCells = Soup1.find("tr", attrs={"class":'a-spacing-small po-number_of_cells'}).find("span", attrs={"class": "a-size-base po-break-word"}).string.strip()
    except AttributeError:
        noOfCells = ""
    return noOfCells

def get_terminal_type(soup):
    try:
        terminalType = Soup1.find("tr", attrs={"class":'a-spacing-small po-terminal_type'}).find("span", attrs={"class": "a-size-base po-break-word"}).string.strip()
    except AttributeError:
        terminalType = ""
    return terminalType

def get_manufacturer(soup):
    try:
        manufacturer = Soup1.find("tr", attrs={"class":'a-spacing-small po-manufacturer'}).find("span", attrs={"class": "a-size-base po-break-word"}).string.strip()
    except AttributeError:
        manufacturer = ""
    return manufacturer

def get_description(soup):
    try:
        heading = Soup1.find("h1", class_="a-size-base-plus a-text-bold")
        list_ = heading.find_next_sibling("ul", class_="a-unordered-list")
        list_items = list_.find_all("li")
        descriptions = [item.text.strip() for item in list_items]
        return descriptions

    except AttributeError:
        description = ""
```

```python
if __name__ == '__main__':
 
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36 Edg/125.0.0.0', 'Accept-Language': 'en-US, en;q=0.5'})

    URL = "https://www.amazon.com/s?k=residential+5kw+5000+watt+lithium+battery+solar+power+system+kit&crid=3GIP6TUUFS2O0&sprefix=5kw+lithium+battery%2Caps%2C430&ref=nb_sb_ss_ts-doa-p_1_19"

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)
    print(webpage)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all("a", attrs={'class':'a-link-normal s-no-outline'})

    # Store the links
    links_list = []

    # Loop for extracting links from Tag Objects
    for link in links:
            links_list.append(link.get('href'))

    d = {"title":[], "price":[], "rating":[], "reviews":[],"availability":[],"brand":[],"noOfCells":[],"terminalType":[],"manufacturer":[],"description":[]}
    
        # Loop for extracting product details from each link 
    for link in links_list:
        new_webpage = requests.get("https://www.amazon.com" + link, headers=HEADERS)

        Soup1 = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['title'].append(get_title(Soup1))
        d['price'].append(get_price(Soup1))
        d['rating'].append(get_rating(Soup1))
        d['reviews'].append(get_review_count(Soup1))
        d['availability'].append(get_availability(Soup1))
        d['brand'].append(get_brand(Soup1))
        d['noOfCells'].append(get_number_of_cells(Soup1))
        d['terminalType'].append(get_terminal_type(Soup1))
        d['manufacturer'].append(get_manufacturer(Soup1))
        d['description'].append(get_description(Soup1))
        

    
    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'].replace('', np.nan, inplace=True)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
